```python
import pandas as pd
# save filepath to variable for easier access
melb_file_path = 'https://raw.githubusercontent.com/neilkava/Datasets/main/melb_data.csv'
# read the data and store data in DataFrame titled melbourne_data
melb_data = pd.read_csv(melb_file_path) 
# print a summary of the data in Melbourne data
melb_data.head()
```

```python
melb_data.columns
```

```python
# The Melbourne data has some missing values (some houses for which some variables weren't recorded.)
# We'll learn to handle missing values in a later tutorial.  
# Your Iowa data doesn't have missing values in the columns you use. 
# So we will take the simplest option for now, and drop houses from our data. 
# Don't worry about this much for now, though the code is:

# dropna drops missing values (think of na as "not available")
melb_data = melb_data.dropna(axis=0)
# axis = 0 to drop the row entry
# axis = 1 to drop the column (property)
melb_data.describe()
```

```python
y = melb_data.Price
melb_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude', 'Longtitude']
X = melb_data[melb_features]
X.describe()
```

```python
y.head()
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melb_model = DecisionTreeRegressor(random_state = 19)

# Fit model
melb_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melb_model.predict(X.head()))
```

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices = melb_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state = 1)
melb_model = DecisionTreeRegressor()
melb_model.fit(train_X, train_y)
val_predictions = melb_model.predict(test_X)
mean_absolute_error(test_y, val_predictions)
```

```python
def get_mae(max_leafs, train_X, test_X, train_y, test_y):
    model = DecisionTreeRegressor(max_leaf_nodes = max_leafs, random_state = 13)
    model.fit(train_X, train_y)
    preds_val = model.predict(test_X)
    mae = mean_absolute_error(test_y, preds_val)
    return(mae)
```

```python
for max_leaf in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf, train_X, test_X, train_y, test_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf, my_mae))
```

# *Random Forest*

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(random_state = 3)
forest_model.fit(train_X, train_y)
new_preds = forest_model.predict(test_X)
print(mean_absolute_error(test_y, new_preds))
```
